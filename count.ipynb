```python
#reverse character #1

A="vaibhav"
rev=" "

for ch in A:
    rev = ch + rev

print(rev)    
```

```python
#reverse character #2

v="mahadev"
rev=" "

for ch in v:
    rev= ch+rev
print(rev)    
```

```python
#shortcut for reverse character

v="vaibhav vaidya"
print(v[::-1])
```

```python
#palindrome or not
V="mom"

if V==V[::-1]:
    print("palindrome")
else:
    print("not palindrome")
```

```python
#REVERSE EACH WORD

n="vaibhav vaidya is very Good boy "
words=n.split()

for ch in words:
    print(ch[::-1])
```
