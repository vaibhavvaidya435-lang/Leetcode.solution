```python
# COUNT A SPECIFIC LETTER

n="vaibhavvaidya"
count=0

for ch in n:
    if ch=="v":
       count=count+1

print(count)
```

```python
#reverse character #1

A="vaibhav"
rev=" "

for ch in A:
    rev = ch + rev

print(rev)    
```

```python
#reverse character #2

v="mahadev"
rev=" "

for ch in v:
    rev= ch+rev
print(rev)    
```

```python
#shortcut for reverse character

v="vaibhav vaidya"
print(v[::-1])
```

```python
#palindrome or not
V="vaibhav"

if V==V[::-1]:
    print("palindrome")
else:
    print("not palindrome")
```

```python
#REVERSE EACH WORD

n="vaibhav vaidya is very Good boy "
words=n.split()

for ch in words:
    print(ch[::-1])
```
