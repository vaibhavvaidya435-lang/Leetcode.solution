# LEVEL 1: Basic Star Patterns

```python
n=4

for i in range(n):
    for j in range(n):
        print("*",end=" ")
    print()        
```

```python
#right angle triangle

s=5
for i in range(s):
    for j in range(i+1):
        print("#",end=" ")
    print(i)
```

```python
#Inverted Triangle

n = 6
for i in range(n):
    for j in range(n - i):
        print("*", end=" ")
    print()
```

```python
#Number Triangle 

n=9
for i in range(n+i):
    for j in range(n-i):
        print(i,end=" ")
    print()    
```

```python
n=9
for i in range(n+i):#when add+1 in both i and j range indexing start from 1
    for j in range(n-i):
        print(i,end=" ")#when add i for print print pattern in same number in line
    print() 
```

```python
#Right Aligned Triangle
#shortcut
n = 4
for i in range(1, n + 1):
    print("  " * (n - i), end="")
    print("* " * i)
```

```python
#Right Aligned Triangle


n = 4
for i in range(1, n + 1):
    for j in range(n - i):
        print("  ", end="")
    for j in range(i):
        print("* ", end="")
    print()
```
