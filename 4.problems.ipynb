```python
#Frequency of characters in a string

input="aabbcccdde"
freq={}

for ch in input:
    if ch in freq:
        freq[ch]=freq[ch]+1
    else:
        freq[ch]=1
print(freq)        


```

```python
s = "aabbccddeee"
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] = -1
    else:
        freq[ch] = 1

found = False

for ch in s:
    if freq[ch] == 1:
        print(ch)
        found = True
        break

if not found:
    print("No non-repeating character")
```
