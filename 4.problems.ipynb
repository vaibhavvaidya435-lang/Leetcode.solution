```python
#Frequency of characters in a string

input="aabbcccdde"
freq={}

for ch in input:
    if ch in freq:
        freq[ch]=freq[ch]+1
    else:
        freq[ch]=1
print(freq)        


```

```python
# Find first non-repeating character


s = "sshrutii"
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] = freq[ch]+1
    else:
        freq[ch] = 1
for ch in s:
    if freq[ch] == 1:
        print(ch)
        break



```

```python
### find from last non  repeating character
s ="aabbccddeeef"
freq={}

for ch in s:
    if ch in freq:
        freq[ch]-=1
    else:
        freq[ch]=1
for ch in s:
    if freq[ch]==1:
        print(ch)
        break            
```
