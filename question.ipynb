```python
# Given a string s, reverse it and print the result.

s = "hello"
rev = ""

for ch in s:
    rev = ch + rev

print(rev)
```

```python
# Given a string s, check whether it is a palindrome or not.

s = "madam"

left = 0
right = len(s) - 1

is_palindrome = True

while left < right:
    if s[left] != s[right]:
        is_palindrome = False
        break
    left += 1
    right -= 1

if is_palindrome:
    print("Palindrome")
else:
    print("Not Palindrome")
```
