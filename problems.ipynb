```python
#Check whether the letter P is present or not.

word = "APPLE"
found = False

for ch in word:
    if ch == "N":
        found = True

if found:
    print("P is present")
else:
    print("P is not present")


```

```python
#Find how many letters are there in the word.

word = "APPLE"
count = 0

for ch in word:
    count = count + 1

print(count)
```

```python
#Given a string, count how many vowels are in it.
Vowels = "aieouAIEOU"
word = "VAIBHAV"
count = 0

for ch in word:
     if ch in Vowels:
         count = count + 1

print(count)
```
