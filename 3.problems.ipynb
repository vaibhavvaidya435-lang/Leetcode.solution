```python
# PRACTICE 1 – Defanging an IP Address

A= "1.1.1.1"
B = ""

for ch in A:
    if ch == ".":
        B = B + "[.]"
    else:
        B = B + ch

print(B)

        

```

```python
# PRACTICE 2 – Replace space


V=" I  LOVE  PYTHON  "
S= " "

for ch in V:
    if ch == " ":
        S= S + "_"
    else:
        S = S + ch

print(S)       
```

```python
# PRACTICE 3 – Replace vowel

A="VAIBHAV"
Vowels="AEIOUaeiou"
B=" "

for ch in A:
    if ch in Vowels:
        B=B+"*"
    else:
        B= B+ch

print(B)       
             
```
